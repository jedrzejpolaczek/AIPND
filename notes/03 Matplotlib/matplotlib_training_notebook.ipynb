# Absolute vs. Relative Frequency

```python
pokemon = pd.read_csv('pokemons.csv')
print(pokemon)
pokemon.head()
```

```python
pkmn_types = pokemon.melt(id_vars = ['Per', 'Pokemon'],
                         value_vars = ['Type_1', 'Type_2'],
                         var_name = 'type_level',
                         value_name = 'type'
                         ).dropna()
pkmn_types
```

```python
type_counts = pkmn_types['type'].value_counts()
type_order = type_counts.index
```

```python
base_color = sb.color_palette()[0]
sb.countplot(data = pkmn_types, y = 'type', color = base_color, order = type_order)
```

```python
n_pokemon = pokemon.shape[0]  # 663 rows is a number of pokemons, 36 columns is the number of... columns!
max_type_count = type_counts[0]  # First value is max value
max_prop = max_type_count / n_pokemon
print(max_prop)
```

```python
tick_props = np.arange(0, max_prop, 0.02)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
```

```python
# Simple trick but now we know percentage propotion of pokemons (more or less...)!
base_color = sb.color_palette()[0]
sb.countplot(data = pkmn_types, y = 'type', color = base_color, order = type_order)
plt.xticks(tick_props * n_pokemon, tick_names)
plt.xlabel('proportion')
```

```python
# A little more accurate approach...
base_color = sb.color_palette()[0]
sb.countplot(data = pkmn_types, y = 'type', color = base_color, order = type_order)

for i in range(type_counts.shape[0]):
    count = type_counts[i]
    pct_string = '{:0.1f}'.format(100*count/n_pokemon)
    plt.text(count+1, i, pct_string, va = 'center')
```

```python
import random

# Generate 100 random numbers between 1 and 6
data = []
for _ in range(0,100):
    n = random.randint(1,100)
    data.append(n)
data_max = 100

plt.figure(figsize = [10, 5])

# left histogram: data plotted in natural units
plt.subplot(1, 2, 1)
bin_edges = np.arange(0, data_max+10, 10)
plt.hist(data, bins = bin_edges)
plt.xlabel('values')

# right histogram: data plotted after direct log transformation
plt.subplot(1, 2, 2)
log_data = np.log10(data) # direct data transform
log_bin_edges = np.arange(0.8, log_data.max()+0.1, 0.1)
plt.hist(log_data, bins = log_bin_edges)
plt.xlabel('log(values)')
```

```python
bin_edges = np.arange(0, data_max+10, 10)
plt.hist(data, bins = bin_edges)
plt.xscale('log')
```

```python
bin_edges = 10 ** np.arange(0.8, np.log10(data_max)+0.1, 0.1)
plt.hist(data, bins = bin_edges)
plt.xscale('log')
tick_locs = [10, 30, 100, 300, 1000, 3000]
plt.xticks(tick_locs, tick_locs)
```

```python
def sqrt_trans(x, inverse = False):
    """ transformation helper function """
    if not inverse:
        return np.sqrt(x)
    else:
        return x ** 2

bin_edges = np.arange(0, sqrt_trans(data_max)+1, 1)
plt.hist(sqrt_trans(data), bins = bin_edges)
tick_locs = np.arange(0, sqrt_trans(data_max)+10, 10)
plt.xticks(tick_locs, sqrt_trans(tick_locs, inverse = True).astype(int))
```

# Kernel Density Estimation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemons.csv')
print(pokemon.shape)
pokemon.head()
```

```python
sb.distplot(pokemon['HP'])
```

```python
data = [0.0, 3.0, 4.5, 8.0]
plt.figure(figsize = [12, 5])

# left plot: showing kde lumps with the default settings
plt.subplot(1, 3, 1)
sb.distplot(pokemon['HP'], hist = False, rug = True, rug_kws = {'color' : 'r'})

# central plot: kde with narrow bandwidth to show individual probability lumps
plt.subplot(1, 3, 2)
sb.distplot(pokemon['HP'], hist = False, rug = True, rug_kws = {'color' : 'r'},
            kde_kws = {'bw' : 1})

# right plot: choosing a different, triangular kernel function (lump shape)
plt.subplot(1, 3, 3)
sb.distplot(pokemon['HP'], hist = False, rug = True, rug_kws = {'color' : 'r'},
            kde_kws = {'bw' : 1.5, 'kernel' : 'tri'})
```

# Line Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemons.csv')
print(pokemon.shape)
pokemon.head()
```

```python
plt.errorbar(data = pokemon, x = 'HP', y = 'Atk')
```

```python
# set bin edges, compute centers
bin_size = 0.25
xbin_edges = np.arange(0.5, pokemon['HP'].max()+bin_size, bin_size)
xbin_centers = (xbin_edges + bin_size/2)[:-1]

# compute statistics in each bin
data_xbins = pd.cut(pokemon['HP'], xbin_edges, right = False, include_lowest = True)
y_means = pokemon['Atk'].groupby(data_xbins).mean()
y_sems = pokemon['Atk'].groupby(data_xbins).sem()

# plot the summarized data
plt.errorbar(x = xbin_centers, y = y_means, yerr = y_sems)
plt.xlabel('HP')
plt.ylabel('Atk')
```

```python
# compute statistics in a rolling window
df_window = pokemon.sort_values('HP').rolling(15)
x_winmean = df_window.mean()['HP']
y_median = df_window.median()['Atk']
y_q1 = df_window.quantile(.25)['Atk']
y_q3 = df_window.quantile(.75)['Atk']

# plot the summarized data
base_color = sb.color_palette()[0]
line_color = sb.color_palette('dark')[0]
plt.scatter(data = pokemon, x = 'HP', y = 'Atk')
plt.errorbar(x = x_winmean, y = y_median, c = line_color)
plt.errorbar(x = x_winmean, y = y_q1, c = line_color, linestyle = '--')
plt.errorbar(x = x_winmean, y = y_q3, c = line_color, linestyle = '--')

plt.xlabel('HP')
plt.ylabel('Atk')
```

```python
bin_edges = np.arange(-3, pokemon['Atk'].max()+1/3, 1/3)
g = sb.FacetGrid(data = pokemon, hue = 'Gen', size = 5)
g.map(plt.hist, "Atk", bins = bin_edges, histtype = 'step')
g.add_legend()
```

```python
def freq_poly(x, bins = 10, **kwargs):
    """ Custom frequency polygon / line plot code. """
    # set bin edges if none or int specified
    if type(bins) == int:
        bins = np.linspace(x.min(), x.max(), bins+1)
    bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2

    # compute counts
    data_bins = pd.cut(x, bins, right = False,
                       include_lowest = True)
    counts = x.groupby(data_bins).count()

    # create plot
    plt.errorbar(x = bin_centers, y = counts, **kwargs)

bin_edges = np.arange(0, pokemon['Atk'].max()+10, 10)
g = sb.FacetGrid(data = pokemon, hue = 'Gen', size = 5)
g.map(freq_poly, "Atk", bins = bin_edges)
g.add_legend()
```
